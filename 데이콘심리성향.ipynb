```
import pandas as pd
from sklearn.model_selection import train_test_split
import lightgbm as lgbm
from sklearn.metrics import confusion_matrix, accuracy_score
```

```
#경로 설정을 꼭 신경써주세요!!
train=pd.read_csv('./train.csv', index_col=0) #인덱스 col=0도 check!
test=pd.read_csv('./test_x.csv', index_col=0)
submission=pd.read_csv('./sample_submission.csv', index_col=0)  
```

```
train
```

```
train_x = train.drop('voted', axis = 1)
train_y = train['voted']
```

```
gender = {'Male': 1,'Female': 2} 
religion ={'Agnostic':1, 'Atheist':2, 'Buddhist':3, 'Christian_Catholic':4, 'Christian_Mormon':5, 'Christian_Protestant':6, 'Christian_Other':7, 'Hindu':8, 'Jewish':9, 'Muslim':10, 'Sikh':11, 'Other':12}
race={'Asian':1, 'Arab':2, 'Black':3, 'Indigenous Australian':4, 'Native American':5, 'White':6, 'Other':7}

```

```
age_group = {'10s': 1,'20s': 2,'30s':3,'40s':4,'50s':5,'60s':6,'+70s':7} 
```

```
train_x.gender = [gender[item] for item in train_x.gender] 
train_x.religion = [religion[item] for item in train_x.religion] 
train_x.race = [race[item] for item in train_x.race] 
```

```
train_x.age_group = [age_group[item] for item in train_x.age_group] 
```

```
test.gender = [gender[item] for item in test.gender] 
test.religion = [religion[item] for item in test.religion] 
test.race = [race[item] for item in test.race] 
test.age_group = [age_group[item] for item in test.age_group] 
```

```
train_x
```

```
model = lgbm.LGBMClassifier(n_estimators=500)
model.fit(train_x, train_y)
```

```
pred_y = model.predict(test)
```

```
pred_y
```

```
submission['voted']=pred_y
```

```
submission
```

```
submission.to_csv('./sample_submission.csv')
```
